```python
import bilby
from bilby.core.result import read_in_result
from matplotlib import pyplot as plt
myresfile = "PSR_B0833-45_result.json"  
res = read_in_result(myresfile)
```

```python
plt.figure()
res.plot_single_density(key='amplitude',file_base_name='PSR_B0833-45_result.json',color='r')
```

```python
import numpy as np
ul = np.quantile(res.posterior["amplitude"], 0.9)
print("The 90% credible upper limit on h0 is {0:.2e}".format(ul))
```

```python
plt.figure()
plt.text(0.5,0.5,'text')
plt.title('The Posterior Distribution on the \nGravitational-Wave Amplitude')
plt.xlabel('xlabel')
plt.ylabel('ylable')
plt.savefig('words.png')
```
